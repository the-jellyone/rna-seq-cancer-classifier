# **1.Dataset Description**

This collection of data is part of the RNA-Seq (HiSeq) PANCAN dataset. It is a random extraction of gene expressions of patients having different types of tumor: BRCA, KIRC, COAD, LUAD, and PRAD. Each sample contains the expression of 20,531 genes for a patient diagnosed with one of the following cancers:

* BRCA-Breast invasive carcinoma (breast cancer),
* KIRC-Kidney renal clear cell carcinoma (kidney),
* COAD-Colon adenocarcinoma (colon),
* LUAD-Lung adenocarcinoma (lung),
* PRAD-Prostate adenocarcinoma (prostate).

```python
import pandas as pd

df = pd.read_csv("/kaggle/input/gene-expression-cancer-rna-seq-donated-on-682016/data.csv")
df.head()
```

```python


X = pd.read_csv("/kaggle/input/gene-expression-cancer-rna-seq-donated-on-682016/data.csv")
y = pd.read_csv("/kaggle/input/gene-expression-cancer-rna-seq-donated-on-682016/labels.csv")
```

```python
print(X.shape)
print(y.shape)
X.head()

```

```python
y.head()
```

```python
X=X.rename(columns={"Unnamed: 0":"sample_id"})
y=y.rename(columns={"Class":"label"})

```

```python
y=y.rename(columns={"Unnamed: 0":"sample_id"})
```

```python
df=X.merge(y, on="sample_id")
df.head()
```

```python
X = df.drop(columns=["sample_id", "label"])
y = df["label"]
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
y_enc=le.fit_transform(y)

print(le.classes_)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y_enc,test_size=0.2,random_state=42
                                                 ,stratify=y_enc)
```

**Gene expression values were standardized to zero mean and unit variance to prevent scale dominance among genes. The dataset was then split into training and test sets using a stratified strategy to preserve class proportions.**

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
X_train_scaled=sc.fit_transform(X_train)
X_test_scaled=sc.fit_transform(X_test)
```

# **2.EDA**

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x=y)
plt.title("Class Distribution")
plt.show()
```

The class distribution shows moderate imbalance across cancer types, with BRCA having the highest number of samples and COAD the lowest. This imbalance motivates the use of robust evaluation metrics such as macro-averaged F1 score.

```python
import numpy as np

feature_variance = np.var(X_train,axis=0)

plt.figure(figsize=(6,4))
plt.hist(feature_variance,bins=100)
plt.title("Feature Variance Distribution(Genes)")
plt.xlabel("Variance(Not scaled)")
plt.ylabel("Count")
plt.show()
```

# **3.Dimensionality Reduction (PCA)**

```python
from sklearn.decomposition import PCA

pca_full = PCA()
pca_full.fit(X_train_scaled)
plt.figure(figsize=(6,4))
plt.plot(np.cumsum(pca_full.explained_variance_ratio_))
plt.xlabel("Number of components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Explained Variance vs PCA components")
plt.grid(True)
plt.show()
```

The plot starts to plateu around the 50 components area, hence we are choosing 50 components for the PCA parameter.


```python
pca_2 = PCA(n_components=2)
X_pca_2 = pca_2.fit_transform(X_train)

plt.figure(figsize=(6,5))
plt.scatter(X_pca_2[:,0], X_pca_2[:,1], c=y_train, cmap="tab10", s=10)
plt.title("PCA (2 components) — Pre-classification")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
pca_2 = PCA(n_components=2)
X_pca_2 = pca_2.fit_transform(X_train_scaled)

plt.figure(figsize=(6,5))
plt.scatter(X_pca_2[:,0], X_pca_2[:,1], c=y_train, cmap="tab10", s=10)
plt.title("PCA (2 components) — Pre-classification")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

Visualization of the raw, unscaled gene expression space shows apparent separability across classes, primarily driven by differences in absolute gene expression magnitudes. However, after standardization, this separability diminishes, indicating that the raw separation is influenced by scale dominance rather than true underlying structure

**This observation motivates the use of dimensionality reduction techniques to capture meaningful relative expression patterns rather than relying on magnitude-driven artifacts.**

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=50, random_state=42)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)

print(pca.explained_variance_ratio_.sum())
```

```python
import numpy as np

plt.figure(figsize=(6,4))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA Explained Variance")
plt.axhline(0.95, linestyle="--")
plt.tight_layout()
plt.show()
```

Principal Component Analysis (PCA) was applied to reduce dimensionality while preserving most of the variance. The cumulative explained variance curve indicates that approximately 50 components capture the majority of the information in the original gene expression space.

```python
X_train_2d = pca.transform(X_train_scaled)[:, :2]
X_test_2d  = pca.transform(X_test_scaled)[:, :2]

plt.figure(figsize=(6,5))
plt.scatter(X_train_2d[:,0], X_train_2d[:,1], alpha=0.5, label="Train")
plt.scatter(X_test_2d[:,0], X_test_2d[:,1], alpha=0.8, label="Test")
plt.legend()
plt.title("Train vs Test Distribution in PCA Space")
plt.tight_layout()
plt.show()
```

# **4. Model Training**

Two ensemble-based classifiers were trained on the PCA-reduced feature space: Random Forest as a baseline model and XGBoost as a more expressive gradient-boosted model. Hyperparameters were selected to balance performance and generalization.

## 4.1 Random Forest


```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

rf=RandomForestClassifier(n_estimators=300,random_state=42,
                          n_jobs=-1,
                          class_weight="balanced")

rf.fit(X_train_pca,y_train)
y_pred = rf.predict(X_test_pca)
```

```python
class_names = le.classes_

print(classification_report(
    y_test,
    y_pred,
    target_names=class_names
))
```

```python
from sklearn.metrics import confusion_matrix


cm_rf = confusion_matrix(y_test, y_pred)

cm_rf_df = pd.DataFrame(
    cm_rf,
    index=class_names,
    columns=class_names
)

plt.figure(figsize=(6,5))
sns.heatmap(cm_rf_df, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix – Random Forest")
plt.ylabel("True Label")
plt.xlabel("Predicted Label")
plt.tight_layout()
plt.show()
```

**Random Forest achieved strong baseline performance, demonstrating the effectiveness of ensemble methods for multiclass cancer classification using reduced transcriptomic features.**

## 4.2 XGBOOST 

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(
    n_estimators=300,
    max_depth=6,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="multi:softprob",
    num_class=len(class_names),
    eval_metric="mlogloss",
    random_state=42
)

xgb.fit(X_train_pca, y_train)
```

```python
y_pred_xgb = xgb.predict(X_test_pca)

print(classification_report(
    y_test,
    y_pred_xgb,
    target_names=class_names
))
```

```python
cm_xgb = confusion_matrix(y_test, y_pred_xgb)

cm_xgb_df = pd.DataFrame(
    cm_xgb,
    index=class_names,
    columns=class_names
)

plt.figure(figsize=(6,5))
sns.heatmap(cm_xgb_df, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix – XGBoost")
plt.ylabel("True Label")
plt.xlabel("Predicted Label")
plt.tight_layout()
plt.show()
```

**XGBoost further improved classification performance, achieving higher accuracy and macro-averaged F1 score, indicating superior modeling of nonlinear decision boundaries.**

```python
rf_f1 = 0.96   # from RF report
xgb_f1 = 0.98  # from XGB report

models = ["Random Forest", "XGBoost"]
f1_scores = [rf_f1, xgb_f1]

plt.figure(figsize=(5,4))
plt.bar(models, f1_scores)
plt.ylabel("Macro F1 Score")
plt.title("Model Performance Comparison")
plt.ylim(0.9, 1.0)
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import learning_curve
import numpy as np
import matplotlib.pyplot as plt

# Random Forest learning curve
train_sizes_rf, train_scores_rf, test_scores_rf = learning_curve(
    rf,
    X_train_pca,
    y_train,
    cv=5,
    scoring="f1_macro",
    n_jobs=-1
)

# XGBoost learning curve
train_sizes_xgb, train_scores_xgb, test_scores_xgb = learning_curve(
    xgb,
    X_train_pca,
    y_train,
    cv=5,
    scoring="f1_macro",
    n_jobs=-1
)

plt.figure(figsize=(7,5))

# RF curves
plt.plot(
    train_sizes_rf,
    np.mean(test_scores_rf, axis=1),
    marker="o",
    label="Random Forest (Validation)"
)

# XGB curves
plt.plot(
    train_sizes_xgb,
    np.mean(test_scores_xgb, axis=1),
    marker="s",
    label="XGBoost (Validation)"
)

plt.xlabel("Training Samples")
plt.ylabel("Macro F1 Score")
plt.title("Learning Curve Comparison: Random Forest vs XGBoost")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**The learning curve comparison shows that XGBoost consistently achieves higher validation macro F1 scores across different training set sizes. Both models benefit from increased data, but XGBoost demonstrates better generalization and faster performance saturation, indicating superior modeling capacity in the reduced feature space.**

# 5.Why macro F1 is the right choice here

we have :
* Multiclass classification
* Class imbalance (BRCA ≫ COAD)
* A biomedical dataset where minority classes matter

**Macro F1:**

* Computes F1 per class
* Then averages them equally
* Penalizes models that perform well only on dominant classes
